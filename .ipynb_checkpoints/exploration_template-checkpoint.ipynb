## Preliminary Wrangling

#### In this project, I will be exploring the Prosper Loan Data set which contains 113,937 loans and 81 variables. <br> I will focus on a subset of data/variables based on my research interests. 

```python
# Import all packages, set up display options and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from datetime import datetime

%matplotlib inline

# Set figure legend font size.
plt.rc('legend',**{'fontsize':12})

# Setting the options below enables Jupyter notebook to display all columns.
pd.set_option('display.max_columns', None)  
pd.set_option('display.expand_frame_repr', False)
pd.set_option('max_colwidth', -1)
```

```python
# Load the dataset.
df = pd.read_csv('prosperLoanData.csv')
```

```python
# high-level overview of data shape and composition
print(df.shape)
df.sample(4)
```

I'm interested in comparing characteristics of Defaulted loans vs non-Defaulted loans, so I first looked into the 'LoanStatus' variable to see how loan status was categorized.

```python
df.LoanStatus.value_counts()
```

As we can see: Loan Status has been characterized into several different categories. For this analysis, I will be focusing on loans that have been completed, charged off and defaulted. Since the Charged off loans also fall under the 'Defaulted' category, I will combine the Charged off loans with Defaulted loans (mark all 'Chargedoff' loans as 'Defaulted').  [More info about various loan status](https://help.bitbond.com/article/20-the-10-loan-status-variants-explained "11 loan status variants explained")

```python
# Subset the dataframe to include only "Completed", "Chargedoff", "Defaulted" loans, save in new variable 'df_sub'. 
df_sub = df.query('LoanStatus in ["Completed", "Chargedoff", "Defaulted"]').copy()
```

```python
# Replace 'Chargedoff' loans with 'Defaulted'.
df_sub.LoanStatus = df_sub.LoanStatus.replace('Chargedoff', 'Defaulted')
```

```python
# Verify the replacement. 
df_sub.LoanStatus.value_counts()
```

```python
# Check for duplicated records
df_sub.duplicated().sum()
```

Fortunately, no duplicated records were found in the dataset. 

```python
# Moving on, let's do a quick check on data types & missing values for the new dataframe 'df_sub'. 
df_sub.info()
```

Interestingly, we see several columns such as `CreditGrade`, `EstimatedEffectiveYield` etc.. have a lot of missing values. After reviewing the [Prosper Loan variable data dictionary](https://docs.google.com/spreadsheets/d/1gDyi_L4UvIrLTEC6Wri5nbaMmkGmLQBk-Yx3z0XDEtI/edit#gid=0 "Prosper Load variable data dictionary"), I discovered the reason. It turns out that Prosper changed their rating system for loans originated after July 2009. Some new features such as 'EstimatedEffectiveYield' and 'EstimatedLoss' were also introduced after July 2009. This generates good opportunity to compare loans originated before/after July 2009. 

Also, I noticed several columns have incorrect data types. As the dataset has 81 columns, I will only focus on wrangling the columns of interest and keep the unused columns for future analysis. 

```python
# First, let's convert the 'ListingCreationDate' column into datetime format.
df_sub['ListingCreationDate'] = pd.to_datetime(df_sub.ListingCreationDate)
```

```python
# Next, convert the 'Term' column into object datatype.
# The reason for this conversion is that I will use the get_text() method for barchart visualization later on.
df_sub['Term'] = df_sub['Term'].astype(str)
```

```python
# Further, convert 'LoanStatus', 'CreditGrade', 'ProsperRating (Alpha)', 'Term', 'IncomeRange' into ordered categorical types.
# I will ignore the 'NC' 'CreditGrade' as I could not find any reliable information about this credit rating. 

ordinal_var_dict = {'LoanStatus': ['Completed', 'Defaulted'],
                    'CreditGrade': ['HR', 'E', 'D', 'C', 'B', 'A', 'AA'],
                    'ProsperRating (Alpha)': ['HR', 'E', 'D', 'C', 'B', 'A', 'AA'],
                    'Term': ['12', '36', '60'],
                    'IncomeRange': ['Not displayed', 'Not employed', '$0', '$1-24,999', 
                                    '$25,000-49,999', '$50,000-74,999', '$75,000-99,999', '$100,000+']}

for var in ordinal_var_dict:
    ordered_var = pd.api.types.CategoricalDtype(ordered = True,
                                                categories = ordinal_var_dict[var])
    df_sub[var] = df_sub[var].astype(ordered_var)
```

```python
# Moving on, I noticed there are missing values for 'DebtToIncomeRatio'. 
# Select the missing values and examine the results.
df_sub[df_sub.DebtToIncomeRatio.isna()]
```

It seems that most of the loans with Null 'DebtToIncomeRatio' also could not verify their income ('IncomeVerifiable' is 'False'). This makes sense because if the borrower's income could not be verified, their 'DebtToIncomeRatio' results may also be unreliable. For the purpose of this analysis, I will drop all records with missing 'DebtToIncomeRatio' as below.

```python
# create a mask and use it to select records with not null 'DebtToIncomeRatio'.
mask = df_sub.DebtToIncomeRatio.notna()
df_sub = df_sub.loc[mask, :]
```

```python
# Further check to see if there are any records with non-varifialble income.
df_sub.IncomeVerifiable.value_counts()
```

```python
# Let's drop these 'False' records as well.
mask2 = df_sub.IncomeVerifiable != False
df_sub = df_sub.loc[mask2, :]
```

```python
# Next, a quick check on summary statistics of numerical columns.
df_sub.describe()
```

I noticed the minimal 'MonthlyLoanPayment' is 0, which doesn't make sense. <br>
From the minimal loan amount of \\$1000, with maximium loan Term of 60 months, even with 0% borrower rate,<br>
the minimal 'MonthlyLoanPayment' should be greater than \\$16.66.

```python
# Query records with 'MonthlyLoanPayment' below and equal to $16.66
df_sub[df_sub.MonthlyLoanPayment <= 16.66]
```

I will drop these invalid records as below.

```python
df_sub = df_sub[df_sub.MonthlyLoanPayment > 16.66]
```

```python
# Let's recheck datatypes & missing values for 'df_sub' dataframe.
df_sub.info()
```

We see that after the data wrangling above, our dataset 'df_sub' has 49640 loan records in total.

```python
# As I wanted to compare loans originated before July 2009 and after July 2009 with their different rating systems,
# I will further subset 'df_sub' into two new dataframes: 'df_old_loans' for loans originated before July 2009,
# and 'df_new_loans' which contains all loans originated after July 2009. 
df_old_loans = df_sub[df_sub['ListingCreationDate'] < '2009-07-01'].copy()
df_new_loans = df_sub[df_sub['ListingCreationDate'] > '2009-07-01'].copy()
```

```python
# A quick check of datatypes & missing values for 'df_old_loans' dataframe.
df_old_loans.info()
```

Since columns: 'EstimatedEffectiveYield', 'EstimatedLoss', 'EstimatedReturn', 'ProsperRating (numeric)', 'ProsperRating (Alpha)', 'ProsperScore' contain info for loans originated after July 2009, it is no surprise they have 0 non-null values (in the 'df_old_loans' dataset). Let's drop these columns as below. 

```python
col_to_drop = ['EstimatedEffectiveYield', 'EstimatedLoss', 'EstimatedReturn', 
               'ProsperRating (numeric)', 'ProsperRating (Alpha)', 'ProsperScore' ]

df_old_loans.drop(col_to_drop, axis=1, inplace=True)
```

## Univariate Exploration

**The first question I have is: What is the overall percentage of defaulted Prosper loans?**

```python
# To answer this question, let's obatain the 'LoanStatus' value counts.
sorted_counts = df_sub.LoanStatus.value_counts(ascending=True)
sorted_counts
```

```python
# Since 'LoanStatus' only has two categories, 
# a pie chart would be a good choice for visualization.  

colors = ['lightcoral', 'lightskyblue']
explode = (0.1, 0)   # only "explode" the 1st slice: 'Defaulted'

plt.figure(figsize=(8, 6), dpi=80)
plt.pie(sorted_counts, explode=explode, labels= sorted_counts.index, colors=colors,
autopct='%1.1f%%', shadow=False, startangle=90, textprops={'fontsize': 14}, counterclock = False)
plt.axis('equal') # Equal aspect ratio will ensure the pie chart is drawn as a circle
plt.title('Prosper Loan Status Distribution', fontsize= 16); 
```

As we can see, the overall default rate for Prosper loans is 30.8% and 69.2% of loans are completed loans. 

**Next we will further explore the 'df_sub' dataset. Question #2: Of all the originated loans, what kind of loan terms are there? <br>**
**How many loans are there in each loan term category?** 

```python
# Let's start with the 'Term' values_counts()
df_sub.Term.value_counts()
```

```python
# Since the loan term has three categories, I will use a barchart for visualization.
# I will also mark the percentage of each category on top of each bar as below.

# Create a barchart
plt.figure(figsize=(7, 4.5), dpi=80)
base_color = sb.color_palette("BrBG", 10)[6]
ax = sb.countplot(data = df_sub, x = 'Term', color = base_color);

n_points = df_sub.shape[0]
term_counts = df_sub.Term.value_counts()
# Obtain current tick locations and labels
locs, labels = plt.xticks()  

# Loop through each location & label pair
for loc, label in zip(locs, labels):

    # Obtain text property for the label  
    count = term_counts[label.get_text()]
    
    pct_string = '{:0.1f}%'.format(100*count/n_points)
    
    # print annotation on top of the barchart bars
    plt.text(loc, count+1000, pct_string, ha = 'center', color = 'black', fontsize = 13);
    
# Set figure title, y= 1.06 adds a little space between the title and the chart.
plt.title('Prosper loan term categories', fontsize= 15.5, y= 1.06)

# Only kept the bottom & left axes line. 
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Define the x & y ticks/labels
plt.xticks(fontsize= 13)
plt.xlabel('Loan Term (months)', fontsize= 13)
plt.yticks(fontsize= 13)
plt.ylabel('count', fontsize= 13);
```

From the above graph, we learned that there are three different loan terms: 12 months, 36 months and 60 months for Prosper loans. <br> 
We also discovered that majority of the loans (90.8%) are 36 month loans.

## Multivariate Exploration

Previously we looked at pairwise relationships between 'LoanStatus' vs 'BorrowerRate', 'LoanStatus' vs 'CreditGrade'/'ProsperRating (Alpha)', and 'BorrowerRate' vs 'CreditGrade'/'ProsperRating (Alpha)'. We found that defaulted loans tend to have higher borrower rates compared to completed loans. We also discovered higher Credit Grades or Prosper Rating (Alpha) tend to have lower borrower rates.**Would there be any differences in borrower rates between completed loans and defaulted loans in each sub-Credit Grade or sub-Prosper Rating (Alpha) category?** In order to find out, we can look at the relationships between 'LoanStatus', 'BorrowerRate' and 'CreditGrade'/'ProsperRating (Alpha)' in a multivariate plot. I will use pointplot to depict the relationships as 'LoanStatus', 'CreditGrade'/'ProsperRating (Alpha)' are categorical features, while 'BorrowerRate' is numeric. 

```python
# Create a pointplot to present the relationship between 'LoanStatus', 'CreditGrade' and 'BorrowerRate'.
plt.figure(figsize=(8, 6.5), dpi=80)
 
ax = sb.pointplot(data = df_sub, x= 'LoanStatus', y= 'BorrowerRate', hue= 'CreditGrade', linestyles="", dodge = 0.3, palette="viridis_r");

# Set figure title, y= 1.05 adds a little space between the title and the chart.
plt.title('Relationship between \n Loan Status, Borrower Rate and Credit Grade', fontsize= 15.5, y= 1.05)

# Only kept the bottom & left axes line. 
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Define the x & y ticks/labels
tick_pos = [0.10, 0.15, 0.20, 0.25]
tick_label = ['{:0.2f}'.format(x) for x in tick_pos]
plt.xticks(fontsize= 13)
plt.xlabel('Loan Staus', fontsize= 13)
plt.yticks(tick_pos,tick_label, fontsize= 13)
plt.ylabel('Mean Borrower Rate', fontsize= 13)
leg = ax.legend(frameon=False)
leg.set_title('Credit Grade',prop={'size':12});
```

We see that for both the completed and defaulted loans, grade 'HR' and grade 'E' loans have the highest mean borrower rate. And as Credit Grade increases in quality, the mean borrower rate tend to be lower (except for 'HR' vs 'E' loans). Interestingly, the defaulted loans have higher mean borrower rate in every Credit Grade compared to the completed loans.
